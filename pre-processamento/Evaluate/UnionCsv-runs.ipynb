```python
dfs = glob.glob('/content/drive/MyDrive/ucf_experiments/ciclo_experimental_1/pre-processamento/evaluate/acc/validation/*.csv')
dfs
```

```python
n_epoch = 30

df_experiments = pd.DataFrame()
for data_path in dfs:
  experiment = data_path.split('/')[-1].split('.')[0].replace('-tag-epoch_categorical_accuracy', '')
  df = pd.read_csv(data_path).head(n_epoch) 
  df['experiment'] = experiment
  df_experiments = pd.concat([df_experiments, df])
```

```python
df_experiments
```

```python
df_experiments.experiment.value_counts()
```

```python
#pd.pivot_table(df_experiments, index=["experiment"], values=["Value"])
stats = pd.pivot(df_experiments, index='Step', columns='experiment', values='Value')
stats
```

```python
from google.colab import files
stats.to_csv('stats.csv') 
files.download("stats.csv")
```
